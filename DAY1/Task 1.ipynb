```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

X = np.array([5,6,7,8,9,10,4,3,7,5,2,9,4,2,3,3 ,7,6]) 
Y = np.array([78,87,80,89,90,34,33,76,50,20,88,76,23,43,31,66,88,30])  

def compute_cost(X, Y, m, b):
    predictions = m * X + b
    cost = np.mean((Y - predictions) ** 2)
    return cost

def gradient_descent(X, Y, learning_rate, max_epochs):
    m, b = 0, 0  
    prev_cost = float('inf')  
    for epoch in range(max_epochs):
        m_gradient = np.mean((m * X + b - Y) * X) 
        b_gradient = np.mean(m * X + b - Y) 
        m -= learning_rate * m_gradient 
        b -= learning_rate * b_gradient  
        current_cost = compute_cost(X, Y, m, b) 
        if current_cost > prev_cost: 
            print("Cost increased! Stopping...")
            break
        prev_cost = current_cost  
    return m, b

m = 0  
b = 0  
learning_rate = 0.01
epochs = 1000
m, b = gradient_descent(X, Y, learning_rate, epochs)
print(f"Optimized Slope (m): {m}")
print(f"Optimized Intercept (b): {b}")

predictions = m * X + b

plt.scatter(X, Y, color='red', label="Actual Data")
plt.plot(X, predictions, color='blue', label="Predicted Line")
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.legend()
plt.title("Linear Regression from Scratch") 
```

```python
#Using Rough Method
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("Pizza.csv")
X = data[["Inches"]].values
Y = data[["Dollar"]].values

m = (sum((X - np.mean(X)) * (Y - np.mean(Y)))) / sum(np.square(X - np.mean(X)))
b = np.mean(Y)-(m * np.mean(X))
y = m * X + b

plt.scatter(X, y, color="blue", label="Actual Data")
plt.plot(X, y, color="red", linewidth=2, label="Regression Line")
plt.xlabel("Study Hours")
plt.ylabel("Scores")
plt.legend()
plt.show()
```
